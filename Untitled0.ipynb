```python
def working_with_data(p):
  df = pd.read_csv(p)
  discrete_values = [0,1]
  df = df.loc[:, df.columns!='Отметка времени']
  df['Пол'] = df['Пол'].replace(['М','Ж'], discrete_values)
  df['Кофе или чай?'] = df['Кофе или чай?'].replace(['Чай','Кофе'],discrete_values)
  df['Занимаетесь спортом?'] = df['Занимаетесь спортом?'].replace(['Да','Нет'],discrete_values)
  df['Есть ли работа?'] = df['Есть ли работа?'].replace(['Да','Нет'],discrete_values)
  df['Есть ли сердечные заболевания?'] = df['Есть ли сердечные заболевания?'].replace(['Да','Нет'],discrete_values)
  df['Сова или Жаворонок'] = df['Сова или Жаворонок'].replace(['Жаворонок','Сова'],discrete_values)
  df['Есть ли молоко в холодильнике?'] = df['Есть ли молоко в холодильнике?'].replace(['Да','Нет'],discrete_values)
  df['Время подъема'] = df['Время подъема'].replace(['5-6 утра','6-7 утра','7-8 утра','Позднее 8 утра'], [0,1,2,3])
  df['Административный округ'] = df['Административный округ'].replace(['Юго-Восточный административный округ',
                                                                       'Южный административный округ',
                                                                       'Восточный административный округ',
                                                                       'Северо-западный административный округ',
                                                                       'Северо-Восточный административный округ',
                                                                       'Юго-Западный административный округ',
                                                                       'Северный административный округ'], [0,1,2,3,4,5,6])
  df['Время сна'] = df['Время сна'].replace(['Менее 3 часов','От 3 до 6 часов',
                                                                       '7-8 часов',
                                                                       'Более 8 часов'],[0,1,2,3])
  return df

import pandas as pd
data = working_with_data('data.csv')
data.head(20)

```
